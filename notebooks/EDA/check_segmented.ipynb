```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
seg_dir = "../../data/segmented"
organs = ['liver','spleen','left_kidney','right_kidney','bowel']
```

```python
def show_plot(organs, slices):
    fig, axes = plt.subplots(1,5,figsize=(20,8))
    for j, (i, organ) in enumerate(zip(slices, organs)):
        path = os.path.join(seg_dir, f'{organ}')
        sample = glob.glob(path + '/*')[0]
        sample = np.load(sample)

        axes[j].imshow(sample[i, 3, :, :], cmap='bone')
        axes[j].imshow(sample[i, -1, :, :], alpha=0.3)
        axes[j].set_title(f"{organ}");
```

```python
show_plot(organs, [8, 8, 8, 8, 16])
```
